```python
# fundamentals
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
shots = pd.read_csv(Path('data')/'NBA_2022_Shots.csv')
shots.head()
```

```python
all_stars = ['LeBron James', 'Kevin Durant', 'Giannis Antetokounmpo', 'Joel Embiid', 'Stephen Curry', 
             'Ja Morant', 'DeMar DeRozan', 'Jayson Tatum', 'Nikola Jokic', 'Andrew Wiggins', 'Trae Young', 
             'Devin Booker', 'Luka Doncic', 'Karl-Anthony Towns', 'Darius Garland', 'Zach LaVine', 
             'Chris Paul', 'Dejounte Murray', 'Jimmy Butler', 'Khris Middleton', 'Donovan Mitchell', 
             'LaMelo Ball', 'Fred VanVleet', 'Rudy Gobert', 'James Harden']
```

```python
stars = shots[shots['PLAYER_NAME'].isin(all_stars)]
stars.head()
```

```python
stars['SHOT_MADE']
```

```python
stars.loc[:, 'SHOT_MADE'] = stars['SHOT_MADE'].apply(lambda x: 1 if x == True else 0)
```

```python
stars_stats = stars.groupby(['PLAYER_NAME', 'BASIC_ZONE'])[['SHOT_MADE']].agg(['sum', 'count', 'mean'])
```

```python
stars_stats.to_json(Path('data')/'allstar_stats.json')
```
