```python
# fundamentals
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
data_dir = Path('data')
vacs_path = data_dir / 'NBA_2022_Shots.csv'
shots = pd.read_csv(vacs_path)
```

```python
shots_count = shots.groupby('PLAYER_NAME').size()

# Filtering out players with more than 500 shots
players_over_500_shots = shots_count[shots_count > 500].index

# Filtering shots DataFrame based on players with more than 500 shots
shots_filtered = shots[shots['PLAYER_NAME'].isin(players_over_500_shots)]
shots_filtered.groupby('PLAYER_NAME').count()
```
