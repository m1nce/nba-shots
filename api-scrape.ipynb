```python
# fundamentals
import pandas as pd
import numpy as np

# web scraping modules
import bs4
import requests
```

```python
# https://www.nba.com/stats/players/shooting?PerMode=Totals&Season=2022-23

response = requests.get('https://www.nba.com/stats/players/shooting?Season=2022-23&PerMode=Totals')
soup = bs4.BeautifulSoup(response.text, features='lxml')
soup
```

```python
soup = (soup
        .find('body')
        .find('div', class_='Layout_base__6IeUC Layout_justNav__2H4H0 Layout_withSubNav__ByKRF')
        .find('div', class_='MaxWidthContainer_mwc__ID5AG')
        .find('section', class_='Block_block__62M07 nba-stats-content-block')
        .find('div', class_='Block_blockContent__6iJ_n')
        .find('div', class_='Crom_container__C45Ti crom-container')
        .find('table', class_='Crom_table__p1iZz')
        .find('tbody', class_='Crom_body__UYOcU'))
soup
```
